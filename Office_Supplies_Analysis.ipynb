## **4. Analysis Operation**

```python
def clean_office_supplies(df):

    df = df.assign(
        OrderDate = pd.to_datetime(df.OrderDate),
        year = pd.to_datetime(df.OrderDate).dt.year,
        month = pd.to_datetime(df.OrderDate).dt.month
    )

    conditions = [
        (df['month'].isin([12, 1, 2])),  # Winter
        (df['month'].isin([3, 4, 5])),   # Spring
        (df['month'].isin([6, 7, 8])),   # Summer
        (df['month'].isin([9, 10, 11]))  # Fall
    ]

    choices = ['Winter', 'Spring', 'Summer', 'Fall']

    df['Season'] = np.select(conditions, choices, default='Unknown')

    df['Total_Profit'] = df['Units'] * df['Unit_Price']

    return df
```

```python
cleaned_office_supplies = clean_office_supplies(office_supplies)
cleaned_office_supplies.head()
```

### **5.3 Time Serie Analysis**

```python
# What are the monthly sales trends over the past year?

monthly_sales = (
    cleaned_office_supplies
    .groupby(['month','year'])
    .Total_Sales
    .sum()
    .reset_index()
)

monthly_sales
```

```python
# Visualization

plt.figure(figsize=(12,6))
sns.lineplot(data= monthly_sales, x='month', y='Total_Sales', marker='o', hue='year', palette='viridis')
plt.title('Monthly Sales Trends Over Past Year')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.legend(title='year')
plt.xticks(range(1,13),['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()
```

>- **Overview:** The analysis revealed varying sales patterns across different months and years, indicating fluctuations in demand for office supplies.

### **Year Wise Insights**
>- **2014:** Sales saw significant increases in July and December above 2500, with noticeable fluctuations throughout the year.
>- **2015:** January, February, and June showed notable increases in sales above 1500, whereas March and April experienced declines.

```python
# Identify any seasonal patterns in the sales data.

seasonal_sales = (
    cleaned_office_supplies
    .groupby('Season')
    .Total_Sales
    .sum()
    .reset_index()
    .sort_values(by='Total_Sales')
)

seasonal_sales
```

>- **Spring :** Spring has the lowest sales among the seasons. This might indicate a period of lower demand.
>- **Fall :** Sales in fall were increased by 30% as comapared to the Highest winter sale and declined sales in Spring. So we have prepare for end of the year and upcoming season.
>- **Winter :** Winter has the highest sales, suggesting increased purchases during this season. This could be due to year-end buying, holiday season demand, or preparations for the new year.
>- **Summer :** Summer follows closely to the winter season in sales.

```python
# What are the sales trends across different regions?

monthly_region_sales = (
    cleaned_office_supplies
    .groupby(['year','month','Region'])
    .Total_Sales
    .sum()
    .reset_index()
    .sort_values(by=['year','month'])
)

monthly_region_sales
```

```python
# Visulaization

regions = monthly_region_sales['Region'].unique()

fig,axes = plt.subplots(nrows=len(regions), ncols=1, figsize=(14, 8),sharex=True)

# Plot the monthly sales trends for each region.
for i, region in enumerate(regions):
    region_data = monthly_region_sales[monthly_region_sales['Region'] == region]
    axes[i].plot(region_data['year'].astype(str) + '-' + region_data['month'].astype(str),
                 region_data['Total_Sales'], marker='o', label=region)
    axes[i].set_title(f'Sales Trend in {region} Region')
    axes[i].set_ylabel('Total Sales')
    axes[i].legend()


fig.suptitle('Monthly Sales Trends by Region', fontsize=16)

# Rotate x-axis labels for all subplots
for ax in axes:
    for label in ax.get_xticklabels():
        label.set_rotation(45)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

#### **Central Region :**
>- **High Sales Period:** Consistently high sales from November 2014 to before March 2015 with totals exceeding 2000. This indicates a strong demand in these months for the Central region.
>- **No Sales Periods:** There are no sales recorded for the entire months of october 2014 and March 2015.

#### **East Region :**
>- **July 2014 Peak:** Sales peaked in July 2014 with totals reaching 2000.
>- **Downward Trend:** A noticeable decline in sales from July 2014 to November 2014.
>- **December 2014 Hike:** There is a moderate increase in sales in December 2014, reaching around 1500.
>- **Decline in 2015:** A decline in sales continues until June 2015.

#### **West Region :**
>- **No Sales in July 2014:** There were no sales recorded in July 2014.
>- **August 2014 Peak:** Sales increased sharply in August 2014, exceeding 800.
>- **September 2014 Downfall:** A significant drop in sales in September 2014.
>- **High Peak in August 2014:** Another sharp increase in sales in August 2014.
>- **Gradual Decrease:** A gradual decrease in sales, forming a slashing pattern, with sales declining up until May 2015 and no sales in June 2015.

### **5.5 Profit Analysis**

>- **Calculate Total Profit with 20% Profit Margin**

>- **We can use another approch like creating feature purchased cost but that will be additional thing and this approach will be helpful when we are building model.**

>- **So we dont have to calculate the Total Profit By 20% of margin beacuse more sales of product will get more profit but at some point some product has less price and some has more so profit will get more for the less Units product thats why here for deep analysis we are going calculate the Total Profit by 20% margin.**

```python
# What is the total profit for each product category?

profitable_products = (
    cleaned_office_supplies
    .assign(
        Total_Profit = cleaned_office_supplies['Total_Sales'] * 0.20
    )
    .groupby('Item')
    .Total_Profit
    .sum()
    .reset_index()
    .sort_values(by='Total_Profit',ascending=False)
)

profitable_products
```

```python
# Visualization

plt.figure(figsize=(16, 4))
bar = plt.bar(profitable_products['Item'], profitable_products['Total_Profit'], color='skyblue')
plt.bar_label(bar)
plt.xticks(rotation=45)
plt.xlabel('Item')
plt.ylabel('Total Profit')
plt.title('Total Profit per Category')
plt.show()
```
